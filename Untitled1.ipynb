```python
import sys
import numpy as py
class FastAreader :
    
    def __init__ (self, fname=''):
        '''contructor: saves attribute fname '''
        
        self.fname = fname
            
    def doOpen (self):
        if self.fname is '':
            return sys.stdin
        else:
            return open(self.fname)
 
    def readFasta (self):
        header = ''
        sequence = ''
        
        with self.doOpen() as fileH:
            header = ''
            sequence = ''
 
            # skip to first fasta header
            line = fileH.readline()
            while not line.startswith('>') :
                line = fileH.readline()
            header = line[1:].rstrip()
 
            for line in fileH:
                if line.startswith ('>'):
                    yield header,sequence
                    header = line[1:].rstrip()
                    sequence = ''
                else :
                    sequence += ''.join(line.rstrip().split()).upper()
        yield header,sequence
```

```python
#!/usr/bin/env python3

########################################################################
# File:program.py
#  executable: program.py
# Purpose:scaffold file for python executables
#   stderr: errors and status
#   stdout:
#          
# Author: David Bernick
# History:      dlb 08/20/2011 Created
#               dlb 08/26/2021 minor updates
########################################################################

########################################################################
# CommandLine
########################################################################
class CommandLine() :
    '''
    Handle the command line, usage and help requests.

    CommandLine uses argparse,
    it implements a standard command line argument parser with various argument options,
    a standard usage and help, and an error termination exception Usage.

    attributes:
    all arguments received from the commandline using .add_argument will be
    available within the .args attribute of an object instantiated from CommandLine.
    For example, if thisCommandLine is an object of the class, and requiredbool was
    set as an option using add_argument, then thisCommandLine.args.requiredbool will
    name that option.
 
    '''
    
    def __init__(self, inOpts=None) :
        '''
        CommandLine constructor.
        
        Implement a parser to interpret the command line argv string using argparse.
        '''
        
        import argparse
        self.parser = argparse.ArgumentParser(description = 'Program prolog - a brief description of what this thing does', 
                                             epilog = 'Program epilog - some other stuff you feel compelled to say', 
                                             add_help = True, #default is True 
                                             prefix_chars = '-', 
                                             usage = '%(prog)s [options] -option1[default] <input >output' 
                                             )
        self.parser.add_argument('-i', '--iterations', type=int, action = 'store')
        self.parser.add_argument('-p', '--psuedocount', type=int, action = 'store')
        self.parser.add_argument('-k', '--length', type=int, action = 'store')
        if inOpts is None :
            self.args = self.parser.parse_args()
        else :
            self.args = self.parser.parse_args(inOpts)
  

class Usage(Exception):
    '''
    Signal a Usage error, evoking a usage statement and eventual exit when raised.
    '''
    def __init__(self, msg):
        self.msg = msg


```

```python
########################################################################
# The Bioinformatics Class
# Here is where most of your solution will end up.
#
########################################################################
class NullModel:
    '''
    Build the thing that needs doing.
    '''
    def __init__(self):
        self.nucleotideProb = {}
        
    
    def prob(self, seqs):
        '''
        '''
        total = ''.join(seqs)
        lenSeqData = len(total)
        nucPr = {}
        for nucleotide in total:
            if nucleotide not in nucPr.keys():
                nucPr[nucleotide] = 1
            else:
                nucPr[nucleotide] += 1
        for nucleotide, count in nucPr.items():
            nucPr[nucleotide] = count/lenSeqData
            
        self.nucleotideProb = nucPr
        return self.nucleotideProb
                
########################################################################
# The Bioinformatics Class
# Here is where most of your solution will end up.
#
########################################################################
class RandomMotif:
    '''
    Build the thing that needs doing.
    '''
    def __init__(self):
        self._motif = ''

    def motif(self, seq, k):
        index = np.random.randint(0, len(seq) - k)
        window = seq[index : index+k]
        
        self._motif = window
        return self._motif
        
        
########################################################################
# The Bioinformatics Class
# Here is where most of your solution will end up.
#
########################################################################
class Profile:
    '''
    Build the thing that needs doing.
    '''
    def __init__(self):
        self.profile = []

    
    def countMatrix (self, seqs, motiflist, k, p):
        matrix = []
        for i in range(k):
            matrix.append({'A': p, 'T': p, 'C': p, 'G': p})
        for index, motif in enumerate(motiflist):
            for i, nucleotide in enumerate(motif):
                matrix[i][nucleotide] += 1
        for row in matrix:
            for nucleotide in row:
                row[nucleotide] = row[nucleotide] / (len(seqs) + (p*4))
        self.profile = matrix
        
        return self.profile
            

            
########################################################################
# The Bioinformatics Class
# Here is where most of your solution will end up.
#
########################################################################
class MotifDict:
    '''
    '''
    def __init__(self):
        self.probs = []
        self.newMotifs = []
        
    def probability (self, seqs, profileMatrix, k):
        prob = []
        newMotifs = []
        #print(6, newMotifs)
        for seq in seqs:
            #print (7, seq)
            dictionary = {}
            windowMax = ('', 0)
            for i in range(len(seq)-k+1):
                window = [seq[i:i+k]][0]
                dictionary[window] = 1
                for j, nucleotide in enumerate(window):
                    dictionary[window] = (dictionary[window] * 
                    profileMatrix[j][nucleotide])
                #print (8, window, dictionary[window])
            
                if dictionary[window] > windowMax[1]:
                    windowMax = (window, dictionary[window])
            #print(9, windowMax)
            newMotifs.append(windowMax[0])
            #print(10, newMotifs)
            prob.append(dictionary)
            #print(dictionary)
            self.newMotifs =newMotifs
        
        return self.newMotifs

########################################################################
# The Bioinformatics Class
# Here is where most of your solution will end up.
#
########################################################################
class MotifFinder:
    
    def __init__(self):
        self.consensus = ''
        self.score = 0
        
    def motifSearch(self, motiflist, seqs, k, p):
            
                
        profileMatrix = Profile().countMatrix(seqs, motiflist, k, p)
        #print(profileMatrix)   
        consensus = ''
        RelativeEntropy = 0
        Q = NullModel().prob(seqs)
        for row in profileMatrix:
            consensus += max(row, key=row.get)
            #print (consensus)
            for nucleotide in row:
                RelativeEntropy -= (row[nucleotide] * np.log2(row[nucleotide]) * (row[nucleotide] / Q[nucleotide]))   
        return (consensus, RelativeEntropy)
    
        
    def iterate(self, seqs, k, p):
        startingMotifs = []
        #print(2, startingMotifs)
        for seq in seqs:
            startingMotifs.append(RandomMotif().motif(seq, k))
        #print(3, startingMotifs)
        x = ('', 0)
        #print(4, x)
        y = self.motifSearch(startingMotifs, seqs, k, p)
        #print(5, y)
        while x[0] != y[0]:
            c = y
            newMotifs = MotifDict().probability(seqs, profileMatrix, k)
            #print(11, newMotifs)
            y = self.motifSearch(newMotifs,seqs, k, p)
            #print (12, y)
            x = c
        #(self.consensus, self.score) = y
        #print(13, x, y)   
        return y
```

```python
########################################################################
# Main
# Here is the main program
# This is a place to instantiate objects, read and write some data for
# those objects and maybe implement some of the command line options.
########################################################################

def main(infile, myCommandLine=None):
    '''
    Get the peptide and establish the mass spectrum for the cycle.

    '''
    
    #myCommandLine = CommandLine(myCommandLine)

    c1 = CommandLine(myCommandLine)
    i = c1.args.iterations
    p = c1.args.psuedocount
    k = c1.args.length
    
    fa = FastAreader(infile)
    
    seqs = []
    for head, seq in fa.readFasta():
        seqs.append(seq.upper())
        #print (1, seq)
    
    max_value = 0
    max_consensus = ''
    for j in range(i):
        outcome = MotifFinder().iterate(seqs, k, p)
        #print(14, outcome)
        if outcome[1] >= max_value:
            max_value = outcome[1]
            max_consensus = outcome[0]
    print(15, max_consensus, max_value)
    
if __name__ == "__main__":
    main('/home/nick_rose/Downloads/CRISPR_test/CONCAT.txt', ['-i 1000', '-p 1', '-k 13'])
```
